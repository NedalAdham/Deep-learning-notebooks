```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(20190585)
```

```python
signal1time = np.linspace( -50 , 200 , 250 ) 
def random(signal1time):
  return np.random.normal(0, 25, len(signal1time))

signal1 = []
for i in range(50):signal1.append(random(signal1time))
signal1 = np.array(signal1)

display(signal1.shape)

plt.figure(figsize=(14, 7))
plt.plot(signal1time * 10 , sum(signal1))
```

```python
signal1mean = np . mean (signal1 , 0)
plt.figure(figsize=(14, 7))
plt.plot(signal1time *10 , signal1mean)
```

```python
baseline_correction =signal1mean -  np.mean (signal1mean[:49])
plt.figure(figsize=(14, 7))
plt.plot( signal1time * 10,baseline_correction)
```

```python
fig, axs = plt.subplots(2 , figsize=(15, 7))
axs[0].plot(signal1time * 10, sum(signal1) )
axs[1].plot( signal1time * 10 ,baseline_correction)
```
